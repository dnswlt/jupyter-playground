# Plotting with matplotlib


Funny magic command to enable inline figures

```python
%matplotlib inline
```

```python
# Standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10)
plt.plot(x, np.sin(x), '--', linewidth=2)
plt.plot(x, np.cos(x), '-')
```

```python
plt.plot(x, np.exp(x), 'yx')
plt.plot(x, np.power(x, 4), 'go')
plt.title('Two plots in one chart')
```

```python
r = np.random.poisson(100, 10000)
h = plt.hist(r, bins=30, color='r')
plt.title('Funny Poisson histogram')
```

```python
N = 100
x = np.random.randn(N)
y = np.random.randn(N)
plt.figure(figsize=(7, 7))   # 7 x 7 inches, whatever 'inch' means in this context
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radiuses
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
```

```python
x = np.linspace(0, 1000)
plt.plot(x, np.random.randn(len(x)).cumsum())
```
